```python
#Library

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
import random as rd
```

```python
#GD Code
n = 20
x = [i for i in range(n)]
y = [i + np.random.rand()*7 for i in range(n)]

w = np.random.rand()
b = np.random.rand()

lr = 0.0005 # learning Rate
iter = 0
maxIter = 1000
loss = []

while True:
    L = 0
    shuffle = list(range(n))
    rd.shuffle(shuffle)
    
    for idx in range(n):
        k = shuffle[idx]
        s = w * x[k] + b*1
        L = L + 1/2 * (s - y[k])**2
        
        #Chain Rule
        dL_ds = s - y[k]
        ds_dw = x[k]
        ds_db = 1
        
        dL_dw = dL_ds * ds_db
        dL_db = dL_ds * ds_dw
        
        w = w - lr*dL_dw
        b = w - lr*dL_db
        
    iter += 1
    loss.append(L)
    y_regress = [w * i + b for i in x]
    
    if iter > maxIter: break
    
#Loss
plt.plot(range(iter), loss, 'b')    
plt.show()

# Prediction
plt.plot(x, y_regress, 'r')
plt.show()
```
