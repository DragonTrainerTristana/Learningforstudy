```python
import pandas as pd 
```

```python
train_label = pd.read_csv("train_label.csv")

train_label
```

```python
train_label = train_label.drop('id',axis = 1)
```

```python
Y = train_label
Y
```

```python
X = pd.read_csv("pca_train.csv")
X = X.iloc[0 :, 0: ]
X
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

sc.fit(X)

X = sc.transform(X)
```

```python
X
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

cld = LinearDiscriminantAnalysis()

cld.fit(X,Y)
y_pred = cld.predict(X)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(Y,y_pred)
```

```python
submit = pd.read_csv("pca_test.csv")

submit = sc.transform(submit)
```

```python
final_submit = cld.predict(submit)
```
