```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
damping = 1.0
inf = pow(10,60)
ppg = 3
w = np.random.randint(0,10,(ppg, ppg))

# matrix
alpha = np.zeros((ppg,ppg))
rho = np.zeros((ppg,ppg))
eta = np.zeros((ppg,ppg))
beta = np.zeros((ppg,ppg))
```

```python
w
```

```python
def update_beta():
    global beta 
    old = np.copy(beta)
    new = w+alpha
    beta = damping*new + (1-damping)*old
    # 
def update_rho():
    global rho 
    old = np.copy(rho)
    
    # 
    # 
    # 
    # 
    
    new = w+eta
    rho = damping*new + (1-damping)*old
    
def update_alpha(): # 행이 보는 입장
    global alpha
    old = np.copy(alpha)
    new = np.copy(alpha)
    for i in range(0,ppg):
        for j in range(0,ppg):
            tmp = np.copy(rho)
            tmp[i,j] = -inf
            new[i,j] = -tmp[i,:].max()
    alpha = damping*new + (1-damping)*old
    
def update_eta(): # 열이 보는 입장
    global eta
    old = np.copy(eta)
    new = np.copy(eta)
    for i in range(0,ppg):
        for j in range(0,ppg):
            tmp = np.copy(beta)
            tmp[i,j] = -inf
            new[i,j] = -tmp[:,j].max()
    eta = damping*new + (1-damping)*old
```

```python
for i in range (0,10):
        
    print("[Iter : {0}] Update Alpha\n".format(i))     
    update_alpha()
    print(alpha)
    print("\n")
    print("[Iter : {0}] Update eta\n".format(i))   
    update_eta()
    print(eta)
    print("\n")    
    print("[Iter : {0}] Update rho\n".format(i))   
    update_rho()
    print(rho)
    print("\n")    
    print("[Iter : {0}] Update beta\n".format(i))   
    update_beta()
    print(beta)
    print("\n")    
```
