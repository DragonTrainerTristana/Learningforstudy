```python
import numpy as np
#np.random.seed(seed=1)

inf = pow(10,60)
damping = 0.5 # ratio of the new values
ppg = 4 #people per group
#w = np.random.randint(1,10,(ppg,ppg))
w = np.array([[8,2,8,1],[7,8,7,2],[1,2,9,9],[4,9,8,4]])
print("w:\n",w)

alpha = np.zeros((ppg,ppg))
rho = np.zeros((ppg,ppg))
eta = np.zeros((ppg,ppg))
beta = np.zeros((ppg,ppg))

def update_rho():
    global rho #variables: local for read-only, global for updating 
    old = np.copy(rho)
    new = np.copy(rho)
    for i in range(0,ppg):
        tmp = np.copy(w+alpha)
        tmp[i,:] = -inf
        for j in range(0,ppg):
            new[i,j] = w[i,j] - tmp.max()
    rho = damping*new + (1-damping)*old
    
def update_beta():
    global beta
    old = np.copy(beta)
    new = np.copy(beta)
    for j in range(0,ppg):
        tmp = np.copy(w+eta)
        tmp[:,j] = -inf
        for i in range (0,ppg):
            new[i,j] = w[i,j] - tmp.max()
    beta = damping*new + (1-damping)*old
    
def update_alpha():
    global alpha
    old = np.copy(alpha)
    new = np.copy(alpha)
    for j in range(0,ppg):
        tmp = np.copy(rho)
        tmp[:,j] = -inf
        for i in range(0,ppg):
            new[i,j] = -tmp.max()
    alpha = damping*new + (1-damping)*old
    
def update_eta():
    global eta
    old = np.copy(eta)
    new = np.copy(eta)
    for i in range(0,ppg):
        tmp = np.copy(beta)
        tmp[i,:] = -inf
        for j in range(0,ppg):
            new[i,j] = -tmp.max()
    eta = damping*new + (1-damping)*old
```

```python
for i in range (0,100):
    update_rho()
    print("rho:\n",rho)
    update_beta()
    print("beta:\n",beta)
    update_alpha()
    print("alpha:\n",alpha)
    update_eta()
    print("eta:\n",eta)
```
